```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\ayush\OneDrive\Desktop\projects\ds_cricket_proj\Battng info.csv")
#removing year from series 
#removing everyone who scored ducks
#only aus and ind
```

```python
# 1. Remove year from 'Series' column (e.g., 'India Tour of Australia 2020' → 'India Tour of Australia')
df['Series'] = df['Series'].str.replace(r'\s*\d{4}$', '', regex=True)
print(df)
```

```python
# 2. Remove rows where Runs is 0 or '0' (duck)
df = df[df['Runs Scored'].astype(str) != '0']
print(df)
```

```python
# 3. Filter only matches involving India or Australia
df = df[df['Match'].str.contains('India') & df['Match'].str.contains('Australia')]
print(df)
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
print(df)
```

```python
df['Series'] = df['Series'].str.replace(r'\d{4}-\d{2}|\d{4}', '', regex=True).str.strip()

```

```python
df.to_csv(r"C:\Users\ayush\OneDrive\Desktop\projects\ds_cricket_proj\Battng info.csv")
```
